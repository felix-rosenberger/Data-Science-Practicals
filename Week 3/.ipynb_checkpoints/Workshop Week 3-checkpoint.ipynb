## Outliers

The text book Section 3.3.3 talks about removing outliers from the data, discarding values that lie a long way from the median or mean value.   For this dataset, they talk about removing any rows where the income is >50K but the age is below 22 or above 72 - basically saying that these values are unlikely to be real.   This is a reasonable thing to do but the code they give for it is buggy.  

They show this formula:

```
df2 = df.drop(df.index[
                      (df.income == ’>50K\n’) &
                      (df[’age’] > df[’age’].median() + 35) & (df[’age’] > df[’age’].median() -15)
])
```

Note that they are dropping from the dataframe `df` any rows where the income is high AND the age is greater than 72 (median + 35) AND the age is greater than 22 (median - 15).   So effectively the last term will have no effect.  
What they really mean is (age > 72) OR (age < 22).  Here are some expressions that express these constraints. I've printed the length of the dataframe after using the boolean expression to select the rows:

````python
print( len( adult[(adult.income == '>50K')] ))
print( len( adult[(adult.age > 72)] ))
print( len( adult[(adult.age < 22)] ))
print( len( adult[(adult.income == '>50K')  & ((adult.age > 72) | (adult.age < 22))] ))
````

Based on these expressions, we can drop rows outside of the target range using the `drop` method as in the textbook.  Note the use of `adult.index`.  The argument to drop is a series giving the indexes of the rows we want to remove from the dataframe.   So we filter the index of the dataframe using the boolean expression to get this.

````python
# remove outliers - people with high income who are too old or too young
adult_clean = adult.drop( adult.index[(adult.income == '>50K') & 
                                      ((adult['age'] > 72) | 
                                       (adult['age'] < 22))]
                        )

# confirm that we have restricted the range of ages for high earners, should see 
adult_clean[adult_clean.income=='>50K'].age.describe()
````

Note that unlike in the textbook, the minimum age is now 22 (not 19).

Given this filtered dataset, repeat the analysis in the remainder of section 3.3.3 to **determine whether the difference between males and females is changed once outliers are removed**. To do this you will need to compute versions of the male and female high earners age series with and without outliers removed.  You can do this directly from the `adult` and `adult_clean` dataframes.

````python
# create "cleaned" dataframes 
males_clean_high = adult_clean.loc[(adult_clean["sex"] == "Male") & (adult_clean["income"] == ">50K")]
females_clean_high = adult_clean.loc[(adult_clean["sex"] == "Female") & (adult_clean["income"] == ">50K")]
````

````python
# Statistics for high income men after removing outliers

males_clean_high_mean = males_clean_high["age"].mean()
males_clean_high_median = males_clean_high["age"].median()
males_clean_high_std = males_clean_high["age"].std()
males_clean_high_min = males_clean_high["age"].min()
males_clean_high_max = males_clean_high["age"].max()

print("High income men statistics:\nMean: ", round(males_clean_high_mean, 2),
     "Median: ", round(males_clean_high_median, 2),
     "Standard Deviation: ", round(males_clean_high_std, 2),
     "Min: ", round(males_clean_high_min, 2),
     "Max: ", round(males_clean_high_max, 2))
````

````python
# Statistics for high income women after removing outliers

females_clean_high_mean = females_clean_high["age"].mean()
females_clean_high_median = females_clean_high["age"].median()
females_clean_high_std = females_clean_high["age"].std()
females_clean_high_min = females_clean_high["age"].min()
females_clean_high_max = females_clean_high["age"].max()

print("High income women statistics:\nMean: ", round(females_clean_high_mean, 2),
     "Median: ", round(females_clean_high_median, 2),
     "Standard Deviation: ", round(females_clean_high_std, 2),
     "Min: ", round(females_clean_high_min, 2),
     "Max: ", round(females_clean_high_max, 2))
````

````python
# Difference between high income men and women after and before outliers are removed

print("Difference after outliers are removed:\nMean: ", round(males_clean_high_mean-females_clean_high_mean, 2),
     "\tMedian: ", round(males_clean_high_median-females_clean_high_median, 2),
     "\tStd. Deviation: ", round(males_clean_high_std-females_clean_high_std, 2), "\n")

print("Difference before outliers are removed:\nMean: ", round(mean_males_high-mean_females_high, 2),
     "\tMedian: ", round(median_males_high-median_females_high, 2),
     "\tStd. Deviation: ", round(STD_males_high-STD_females_high, 2))
````

Generate figure 3.5 from the textbook that shows the relative frequency of male an female high earners as a function of age.  Again, label your axes and the plot overall.

````python
# create two histograms

countx,divisionx = np.histogram(males_clean_high["age"], density=True)
county,divisiony = np.histogram(females_clean_high["age"], density=True)
````

````python
# create plot of relative frequency of male and female high earners as a function of age

val = [(divisionx[i]+divisiony[i])/2 for i in range(len(divisionx)-1)] #list comprehension in one line
plt.plot(val, countx-county, 'o-') 
plt.title('Differences in promoting men vs. women')
plt.xlabel('Age',fontsize=15)
plt.ylabel('Differences',fontsize=15)
plt.show()

# -1 to remove outlier
````

Finally, use the **Pearson’s median skewness coefficient** from Section 3.3.4 to confirm the difference in the shape of the distribution of male and female age distributions. 

````python
# Pearson's median skewness coefficient

def pearson(x):
    return 3*(x.mean() - x.median())*x.std()

print("Pearson median skewness coefficient for the male population = ", pearson(males_clean_high["age"]))
print("Pearson median skewness coefficient for the female population = ", pearson(females_clean_high["age"]))
````

### Does Education Level have an Effect on Income?

````python
# first overview over DataFrame again
adult.head()
````

````python
# create dataframe with subgroups

educ_group = adult.groupby(by=["education"])["income"].value_counts()
educ_group_unstacked = educ_group.unstack()
educ_group_unstacked
````

````python
# create dataframe with percentages for subgroups

educ_fractions = adult.groupby(by=["education"])["income"].value_counts(normalize=True).unstack()*100
educ_fractions
````

````python
# concatenate both and rename columns

educ_concat = pd.concat([educ_group_unstacked, educ_fractions], axis=1)
educ_concat.columns = ['<=50K', '>50K', '% <=50K', '% >50K']
educ_concat.sort_values('% >50K', ascending=False)
````

It can clearly be seen that a low education level correlates with low income. The fraction of high income earners increases with higher education levels, so the assumption that higher education leads to higher income is sufficient.

### Are married people more or less highly educated than those who are single?

````python
# first identify unique values to distinguish between married and single

adult["marital"].unique()
````

From here we can categorize: <br />
Married: 'Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse' <br />
Single: 'Never-married', 'Divorced', 'Separated', 'Widowed'

````python
# create new column married with yes or no

adult["Married"] = np.where(adult["marital"] == 'Married-civ-spouse', True, 
                            (np.where(adult["marital"] == 'Married-spouse-absent', True,
                                     (np.where(adult["marital"] == 'Married-AF-spouse', True, False)))))
adult.head()
````

````python
# group by marital status and calculate average education_num for each group

marry_group = adult.groupby(by=["Married"])["education_num"].mean()
marry_group
````

In this sample, married people tend to have, on average, a slightly higher education level.
